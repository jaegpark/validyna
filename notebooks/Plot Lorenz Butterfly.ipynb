```python
import matplotlib.pyplot as plt
from dysts.flows import Lorenz
```

```python
attractor = Lorenz()
traj = attractor.make_trajectory(n=500000)
traj.shape
```

```python
fig = plt.figure(figsize=(9, 8))
ax = plt.axes(projection ='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot(xs=traj[:, 0], ys=traj[:, 1], zs=traj[:, 2])

fig.tight_layout()
fig.subplots_adjust(left=-0.05, right=1, bottom=-0.05, top=1.08, hspace=0)
fig.savefig('Lorenz Butterfly.pdf', format='pdf');
```
